#As a first step, we'll exclude the last 19 columns in the DataFrame due to their containing NaN values, making them unsuitable for # # #further analysis.

```python
df.fillna(0,inplace=True)
df.tail()
```

```python
new_df = df.iloc[:, :3]
new_df
```

```python
#Here we will Merge the different subjects dataframes to create the required dataframe
```

```python
df_filtered_tab = df.loc[:, df.isin(['Tableau']).any()]
df_filtered_dvsr = df.loc[:, df.isin(['DVSR']).any()]
df_filtered_dadm = df.loc[:, df.isin(['DADM']).any()]
df_filtered_nlp = df.loc[:, df.isin(['NLP']).any()]
```

```python
df1 = pd.concat([new_df,df_filtered_tab, df_filtered_dvsr, df_filtered_dadm, df_filtered_nlp], axis=1)
df1
```

```python
#In the dataset, certain columns are currently labeled as "unnamed.
#"To address this, we aim to set the first row as the header, eliminating the current row and improving column identification.
```

```python
df1.columns = df1.iloc[0]
df1 = df1.drop(0)
df1=df1.reset_index(drop=True)
df1
```

```python
filtered_columns_tab = df1.loc[:, df1.columns.str.contains("Tableau")]
filtered_columns_dvsr = df1.loc[:, df1.columns.str.contains("DVSR")]
filtered_columns_dadm = df1.loc[:, df1.columns.str.contains("DADM")]
filtered_columns_nlp = df1.loc[:, df1.columns.str.contains("NLP")]
new_df1 = df1.iloc[:, :3]
df_tab = pd.concat([new_df1,filtered_columns_tab],axis=1)
df_dvsr = pd.concat([new_df1,filtered_columns_dvsr],axis=1)
df_dadm = pd.concat([new_df1,filtered_columns_dadm],axis=1)
df_nlp = pd.concat([new_df1,filtered_columns_nlp],axis=1)
```

```python
#Used for creating total classes column for each subject
df1['Total classes'] = df1.shape[1]-3
df_dvsr['Total classes'] = df_dvsr.shape[1]-3
df_tab['Total classes'] = df_tab.shape[1]-3
df_dadm['Total classes'] = df_dadm.shape[1]-3
df_nlp['Total classes'] = df_nlp.shape[1]-3
```

```python
df1.iloc[:, 3:] = df1.iloc[:, 3:].apply(pd.to_numeric, errors='coerce', downcast='integer')
df_dvsr.iloc[:, 3:] = df_dvsr.iloc[:, 3:].apply(pd.to_numeric, errors='coerce', downcast='integer')
df_tab.iloc[:, 3:] = df_tab.iloc[:, 3:].apply(pd.to_numeric, errors='coerce', downcast='integer')
df_dadm.iloc[:, 3:] = df_dadm.iloc[:, 3:].apply(pd.to_numeric, errors='coerce', downcast='integer')
df_nlp.iloc[:, 3:] = df_nlp.iloc[:, 3:].apply(pd.to_numeric, errors='coerce', downcast='integer')
```

```python
df1['Present'] = df1.iloc[:,3:129].sum(axis = 1)
df_dvsr['Present'] = df_dvsr.iloc[:,3:31].sum(axis = 1)
df_dadm['Present'] = df_dadm.iloc[:,3:51].sum(axis = 1)
df_nlp['Present'] = df_nlp.iloc[:,3:21].sum(axis = 1)
df_tab['Present'] = df_tab.iloc[:,3:35].sum(axis = 1)
```

```python
df1['Percentage'] = (df1['Present']/df1['Total classes'])*100
df_dvsr['Percentage'] = (df_dvsr['Present']/df_dvsr['Total classes'])*100
df_tab['Percentage'] = (df_tab['Present']/df_tab['Total classes'])*100
df_nlp['Percentage'] = (df_nlp['Present']/df_nlp['Total classes'])*100
df_dadm['Percentage'] = (df_dadm['Present']/df_dadm['Total classes'])*100
```

```python
import plotly.graph_objects as go

# Assuming df is your DataFrame
heatmap_data = df1[['Tableau', 'DVSR', 'DADM', 'NLP', 'Total classes']].corr()

fig = go.Figure(data=go.Heatmap(
    z=heatmap_data.values,
    x=heatmap_data.columns,
    y=heatmap_data.index,
    colorscale='Viridis',
    text=heatmap_data.values.round(2),
    hoverinfo='text'
))

fig.update_layout(
    title='Correlation Heatmap for Numeric Columns',
    xaxis=dict(title='Total classes'),
    yaxis=dict(title='Percentage'),
)

fig.show()
```

```python
import plotly.subplots as sp
import plotly.graph_objects as go

# Assuming df is your DataFrame
# Assuming your columns are 'Stu Name', 'Subject', 'Percentage'

# Create subplots
fig = sp.make_subplots(rows=2, cols=1, subplot_titles=['Bar Chart of Percentage', 'Line Chart of Percentage'])

# Bar Chart
bar_chart = go.Bar(
    x=df1['Percentage'],
    y=df1['Stu Name'],
    text=df1['NLP'],
    marker=dict(color='blue'),
    name='Bar Chart'
)

# Line Chart
line_chart = go.Scatter(
    x=df1['Percentage'],
    y=df1['Stu Name'],
    mode='lines+markers',
    text=df1['DADM'],
    marker=dict(color='red'),
    name='Line Chart'
)

# Add traces to subplots
fig.add_trace(bar_chart, row=1, col=1)
fig.add_trace(line_chart, row=2, col=1)

# Update layout
fig.update_layout(
    title='Percentage of Each Student by Subject',
    xaxis=dict(title='Student Name'),
    yaxis=dict(title='Percentage'),
    showlegend=False  # To hide the legend, as it's the same for both charts
)

# Show the plot
fig.show()
```
